```python
import pandas as pd
```

```python
df_manoutput = pd.read_excel(r'C:\Users\adinr\Downloads\Python_Project-1_ESDM\Python_Project-1_ESDM\Data\indonesia-manufacturing-output (1).xlsx')
df2_pngprod = pd.read_excel(r'C:\Users\adinr\Downloads\Python_Project-1_ESDM\Python_Project-1_ESDM\Data\Petroleum and Natural Gas Production, 1996-2022.xlsx')
df3_gdp = pd.read_excel(r'C:\Users\adinr\Downloads\Python_Project-1_ESDM\Python_Project-1_ESDM\Data\GDP.xlsx')
df_employment = pd.read_excel(r'C:\Users\adinr\Downloads\Python_Project-1_ESDM\Python_Project-1_ESDM\Data\Population 15 Years To Top Who Worked by Main Industry 1986-2024.xlsx')
```

```python
df_manoutput['Date'] = pd.to_datetime(df_manoutput['Date'], format='%mm/%dd/%Y')
df_manoutput['Year'] = df_manoutput['Date'].dt.year
df_manoutput.rename(columns={' GDP (Billions of US $)': 'Billions of US $'}, inplace=True)

df_manoutput
```

```python
df_manoutput.rename(columns={
    'Billions of US $': 'Manufacture_Output_Billions_of_USD',
}, inplace=True)
df_manoutput
```

```python
df2_pngprod.rename(columns={'Tahun': 'Year', 'Natural Gas\n(MMscf)':'Natural_Gas_Output_MMscf'}, inplace=True)
df2_pngprod
```

```python
df3_gdp_long = df3_gdp.melt(id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'], 
                          var_name='Year', value_name='GDP')
df3_gdp_long
```

```python
df3_gdp_long.dropna(subset=['GDP'], inplace=True)
df3_gdp_long['Year'] = df3_gdp_long['Year'].astype(int)
```

```python
df_manoutput = df_manoutput[['Year', 'Manufacture_Output_Billions_of_USD']]
df2_pngprod = df2_pngprod[['Year', 'Natural_Gas_Output_MMscf']]
```

```python
# Melt the DataFrame to long format
df_employment_long = df_employment.melt(id_vars=['Lapangan Pekerjaan Utama'], 
                                        var_name='Year', value_name='Employment')
df_employment_long['Year'] = df_employment_long['Year'].astype(int)
# Print the cleaned employment data
print(df_employment_long.head())
```
